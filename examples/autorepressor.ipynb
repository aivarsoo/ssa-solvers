```python
import numpy as np
import torch
import torch.nn.functional as F
import scipy.stats as st
import scipy.interpolate as interpolate
import matplotlib.pyplot as plt

%load_ext line_profiler
%matplotlib inline

torch.set_default_tensor_type(torch.FloatTensor) # no need for double precision 
device = torch.device('cuda:0') if torch.cuda.is_available else torch.device("cpu")
# device = torch.device("cpu")
os.chdir('/home/aivar/projects/ssa-solvers')
```

```python
import os, sys 
from ssa_solvers.simulators import StochasticSimulator, DeterministicSimulator 
from ssa_solvers.data_class import SimulationData

end_time = 400
n_steps = 150 # recommend using 150 - 200 points
n_traj = 10000
time_grid = np.arange(0, end_time, int(end_time / n_steps))
```

### Simulating in cis TetR autorepressor with sRNA  circuit

```python
from circuits.auto_repressor.tetr_srna_incis import TetRsRNAInCis, cfg

data_set_incis = SimulationData(save_to_file=True, device=device)
reaction_system_incis = TetRsRNAInCis(device=torch.device("cpu"))
ode_simulator = DeterministicSimulator(
    reaction_system=reaction_system_incis,
    cfg=cfg
)
reaction_system_incis = TetRsRNAInCis(device=device)
ssa_simulator_incis = StochasticSimulator(
    reaction_system=reaction_system_incis,
    cfg=cfg,
    device=device
)
reaction_system_incis.params = {'aTc': 100}  # setting aTc level
init_pops = torch.zeros((reaction_system_incis.n_species, ), dtype=torch.int64, device=device) #torch.randint(1, (reaction_system.n_species, ))
ssa_simulator_incis.simulate(init_pops=init_pops, end_time=end_time, n_trajectories=n_traj)
ode_res_incis = ode_simulator.simulate(init_pops=np.zeros((reaction_system_incis.n_species,)), time_grid=time_grid)
```

### Simulating in trans TetR autorepressor with sRNA  circuit

```python
from circuits.auto_repressor.tetr_srna_intrans import TetRsRNAInTrans, cfg

data_set_intrans = SimulationData(save_to_file=True, device=device)
reaction_system_intrans = TetRsRNAInTrans(device=torch.device("cpu"))
ode_simulator = DeterministicSimulator(
    reaction_system=reaction_system_intrans,
    cfg=cfg
)
reaction_system_intrans = TetRsRNAInTrans(device=device)
ssa_simulator_intrans = StochasticSimulator(
    reaction_system=reaction_system_intrans,
    cfg=cfg,
    device=device
)
reaction_system_intrans.params = {'aTc': 100}  # setting aTc level
init_pops = torch.zeros((reaction_system_intrans.n_species, ), dtype=torch.int64, device=device) #torch.randint(1, (reaction_system.n_species, ))
ssa_simulator_intrans.simulate(init_pops=init_pops, end_time=end_time, n_trajectories=n_traj)
ode_res_intrans = ode_simulator.simulate(init_pops=np.zeros((reaction_system_intrans.n_species,)), time_grid=time_grid)
```

### Comptuting the statistics on a grid

```python
means_incis = ssa_simulator_incis.data_set.mean(time_range=time_grid)
stds_incis = ssa_simulator_incis.data_set.std(time_range=time_grid)
cov_incis = ssa_simulator_incis.data_set.coefficient_of_variation(time_range=time_grid)

means_intrans = ssa_simulator_intrans.data_set.mean(time_range=time_grid)
stds_intrans = ssa_simulator_intrans.data_set.std(time_range=time_grid)
cov_intrans = ssa_simulator_intrans.data_set.coefficient_of_variation(time_range=time_grid)
```
